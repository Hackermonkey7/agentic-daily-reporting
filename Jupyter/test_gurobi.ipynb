```python
import gurobipy as gp
from gurobipy import GRB

m = gp.Model("test_nb")
x = m.addVar(vtype=GRB.BINARY, name="x")
y = m.addVar(vtype=GRB.BINARY, name="y")
m.setObjective(x + y, GRB.MAXIMIZE)
m.addConstr(x + 2*y <= 2)
m.optimize()

for v in m.getVars():
    print(v.varName, v.x)
print("Obj:", m.objVal)
```
